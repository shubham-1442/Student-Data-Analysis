# GENDER DISTRIBUTION

```python
plt.figure(figsize=(5,5))
random=sns.countplot(data=df,x="Gender",color="steelblue")
plt.title("GENDER DISTRIBUTION")
random.bar_label(random.containers[0])
plt.show()
```

# From the above chart, it is clearly evident that the number of females in the dataset is higher than the number of males

```python
groupbystudent= df.groupby("ParentEduc").agg({"MathScore":"mean", "ReadingScore":"mean","WritingScore":"mean"})
print(groupbystudent)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(groupbystudent,annot=True)
plt.title("PARENT's EDUCATION")
plt.show()
```

# In conclusion, the data presented in the chart underscores the significant influence of parental education on student's academic performance.

```python
groupbymarriage= df.groupby("ParentMaritalStatus").agg({"MathScore":"mean", "ReadingScore":"mean","WritingScore":"mean"})
print(groupbymarriage)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(groupbymarriage,annot=True)
plt.title("PARENT'S MARITAL STATUS")
plt.show()
```

# From the above chart, it is clearly evident that the marital status of parents has a negligible impact on students' scores.

# Distribution of Ethnic Group

```python
groupA=df.loc[(df['EthnicGroup']=="group A")].count()
groupB=df.loc[(df['EthnicGroup']=="group B")].count()
groupC=df.loc[(df['EthnicGroup']=="group C")].count()
groupD=df.loc[(df['EthnicGroup']=="group D")].count()
groupE=df.loc[(df['EthnicGroup']=="group E")].count()
#Creating a list for Pie Chart
m1list=[groupA['EthnicGroup'],groupB['EthnicGroup'],groupC['EthnicGroup'],groupD['EthnicGroup'],groupE['EthnicGroup']]
l=["group A","group B","group C","group D","group E"]
plt.pie(m1list,labels=l,autopct= "%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```
